```python
import zipfile

zip_ref = zipfile.ZipFile("data_object_calib.zip", 'r')
zip_ref.extractall("kitti_data")
zip_ref.close()

print("KITTI dataset extracted")
```

```python
# Sensor range metadata (example values in meters)

sensors = {
    "Camera": 80,
    "Radar": 150,
    "LiDAR": 200,
    "Ultrasonic": 10
}

# select sensor with maximum range
selected_sensor = max(sensors, key=sensors.get)

print("Selected Sensor:", selected_sensor)
```
